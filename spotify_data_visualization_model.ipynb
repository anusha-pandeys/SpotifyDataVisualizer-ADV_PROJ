```python
import math

import pandas as pd
organized_data = pd.DataFrame(columns=['endTime', 'artistName', 'trackName', 'msPlayed'])

def substring_after(s, delim):
    return s.partition(delim)[2]

val = ""

i = 0
j = 0


for index, row in data_df.iterrows():
    j += 1

    k = i/4
    k = math.floor(k)

    val = row.to_string()

    if val.find("endTime") != -1:
        value = substring_after(val, " : \"")
        value = value[:-2]
        #organized_data.iloc[k, organized_data.columns.get_loc('endTime')] = value
        #organized_data.loc[k] = [value, None, None, None]
        #organized_data['endTime'] = organized_data['endTime'].append(pd.Series(value))
        #organized_data.loc[len(organized_data)] = pd.Series({'endTime': value})
        organized_data.loc[k, 'endTime'] = value
        i += 1

    if val.find("artistName") != -1:
        value = substring_after(val, " : \"")
        value = value[:-2]
        organized_data.iloc[k, organized_data.columns.get_loc('artistName')] = value
        #organized_data.loc[k] = [None, value, None, None]
        #organized_data['artistName'] = organized_data['artistName'].append(pd.Series(value))
        #organized_data.loc[len(organized_data)] = pd.Series({'artistName': value})
        organized_data.loc[k, 'artistName'] = value
        i += 1

    if val.find("trackName") != -1:
        value = substring_after(val, " : \"")
        value = value[:-2]
        organized_data.iloc[k, organized_data.columns.get_loc('trackName')] = value
        #organized_data.loc[k] = [None, None, value, None]
        #organized_data['trackName'] = organized_data['trackName'].append(pd.Series(value))
        #organized_data.loc[len(organized_data)] = pd.Series({'trackName': value})
        organized_data.loc[k, 'trackName'] = value
        i += 1

    if val.find("msPlayed") != -1:
        value = substring_after(val, " : ")
        organized_data.iloc[k, organized_data.columns.get_loc('msPlayed')] = value
        #organized_data.loc[k] = [None, None, None, value]
        #organized_data['msPlayed'] = organized_data['msPlayed'].append(pd.Series(value))
        #organized_data.loc[len(organized_data)] = pd.Series({'msPlayed': value})
        organized_data.loc[k, 'msPlayed'] = value
        i += 1

    if j == 100:
      break
```

make new df most_played_songs, w/ columns "trackName," "artistName," "totMsPlayed," "ranking." "ranking" column holds song rank, indicating how often a song was played (1 = most played, large number = least played). use the "msPlayed" value in organized_data to find most played songs. also, some songs in organized_data can be repeats, so add the "msPlayed" values for the repeated songs before inserting in most_played_songs df.


```python
#top/most played songs model

import pandas as pd
most_played_songs = pd.DataFrame(columns=['trackName', 'artistName', 'totMsPlayed', 'ranking'])

song_count = organized_data['trackName'].value_counts()
song_count_dict = song_count.to_dict()

i = 0
j = 1

for key, value in song_count_dict.items():
    artist_name = organized_data[organized_data['trackName'] == key]['artistName'].iloc[0]
    ms_played = 0
    for index, row in organized_data[organized_data['trackName'] == key].iterrows():
        ms_played += float(row['msPlayed'])
    most_played_songs.loc[i, 'trackName'] = key
    most_played_songs.loc[i, 'artistName'] = artist_name
    most_played_songs.loc[i, 'totMsPlayed'] = ms_played
    most_played_songs.loc[i, 'ranking'] = j
    j += 1
    i += 1

most_played_songs = most_played_songs.sort_values(by=['totMsPlayed'], ascending=False)

most_played_songs
```
